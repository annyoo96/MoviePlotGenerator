```python
result = pd.concat([new_kaggle,CMU],axis=0)
```

```python
result
```

```python
result.sort_values("Title", inplace = True) 
dropped_data = result.drop_duplicates(['Title'], keep='first')
#dropped_data.drop_duplicates(['Plot'], keep='first')
```

```python
dropped_data
```

```python
new_data = dropped_data.drop_duplicates(['Plot'], keep='first')
new_data.loc[new_data['Genre'].isnull(),['Genre']] = new_data.loc[new_data['Genre'].isnull(),'Genre'].apply(lambda x: [])
new_data['Genre'] = new_data['Genre'].apply(lambda x: list(set(x)))

```

```python
new_data
```

```python
new_data = new_data.sample(frac=1)
new_data
```

```python
selection = ['horror']

horror = new_data.Genre.apply(lambda x: any(item for item in selection if item in x))
genre_horror = new_data[horror]
genre_horror

horror_test_data = genre_horror.iloc[:300]
horror_train_data = genre_horror.iloc[300:]

horror_test_data.to_csv("Genre_horror_test.csv")
horror_train_data.to_csv("Genre_horror_train.csv")

horror_test_data['Plot'].to_csv('horror_test_plot.txt', sep='\n', index=False)
horror_train_data['Plot'].to_csv('horror_train_plot.txt', sep='\n', index=False)

#genre_horror.to_csv("Genre_horror.csv")
#genre_horror['Plot'].to_csv('Genre_horror_plot.txt', sep='\n', index=False)
```
